Import pandas

```python
import pandas as pd
```

Read the combine.csv file as a DataFrame and investigate the contents

```python
data = pd.read_csv('combine.csv', encoding='latin1')
data
```

Fill missing values for college with 'No College'

```python
import numpy as np

mapping = {np.nan: 'No College'}
data.replace({'college': mapping}, inplace=True)
data.college
```

Remove the players that have null values for the pick

```python
data = data.dropna(how='all', subset=['pick'])
data.pick
```

Investigate the unique values in the position column

```python
pd.unique(data['position'])
```

Replace RB and QB with Running Back and Quarterback

```python
mapping = {'RB': 'Running Back', 'QB': 'Quarterback'}
data.replace({'position': mapping}, inplace=True)
data['position']
```

Create dummy values for position

Merge the dummy data with the original data set

Convert weight from lbs to kg

```python
data.loc[:, 'weight'] = data.weight * 0.45
```

Capitalize name

```python
data.rename(columns={'name':'Name'}, inplace=True)
```

Reverse order of first and last name

```python
columns = data.columns.tolist()
print(columns);

columns = columns[:2] + columns[3:4] + columns[2:3] + columns[4:]

data = data[columns]
data
```
